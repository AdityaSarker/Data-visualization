```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("F:\Data_Visualization_with_Python-main\Data_Visualization_with_Python-main/ford.csv")
display(df)
```

```python
sns.scatterplot(data=df, x='price', y='mileage', hue='fuelType')
```

```python
#Scatterplot of price vs mileage is shown above. fuel type is added as hue.
```

```python
plt.figure(figsize=(20,15))

sns.histplot(data=df,x='mileage',binwidth=2000)
```

```python
#Histogram of price vs mileage is shown above
```

```python
plt.figure(figsize=(20,15))
sns.histplot(data=df,x='mileage',binwidth=200, kde='true')
```

```python
#distribution of mileage is shown above.
```

```python
plt.figure(figsize=(20,15))
sns.histplot(data=df,x='mileage',binwidth=50, cumulative='true')
```

```python
#ECDF of mileage is shown above
```

```python
plt.figure(figsize=(20,15))
plt.subplot(2,3,1)
sns.histplot(data=df,x='mileage',binwidth=2000)
plt.subplot(2,3,2)
sns.histplot(data=df,x='mileage',binwidth=50, cumulative='true')
plt.subplot(2,3,3)
sns.histplot(data=df,x='mileage',binwidth=2000, kde='true')
plt.tight_layout()
plt.show()
```

```python
# Subplot of Histogram, ECDF and distribution showing the count of mileage is shown above.
```
